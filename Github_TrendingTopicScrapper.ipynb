```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def get_topics_page():
    
    topics_url = 'https://github.com/trending'
    response = requests.get(topics_url)
    if response.status_code != 200:
        raise Exception('Failed to load page {}'.format(topics_url))
    doc = BeautifulSoup(response.text, 'html.parser')
    return doc
```

```python
def scrape_username_topic_url(doc):
  url=[]
  username=[]
  topic=[]
  topic_desc=[]

  topic_url_tag=doc.find_all('h1',{'class':'h3 lh-condensed'})
  topic_desc_tag=soup.find_all('p',{'class':"col-9 color-fg-muted my-1 pr-4"})
  for i,j in zip(topic_url_tag,range(len(topic_desc_tag))):
    temp=i.find("a", recursive=False).text.strip().replace("\n", "").split('/')
    username.append(temp[0].strip())
    topic.append(temp[1].strip())
    url.append("https://github.com"+i.find("a", recursive=False)['href'])
    topic_desc.append(topic_desc_tag[j].text.strip())
  #print(topic_desc)

  return [username,topic, url,topic_desc]
```

```python
def scrape_prog_fork_stars_starstoday(doc):
  programming_lang=[]
  stars=[]
  fork=[]
  stars_today=[]


  programming_lang_tag=doc.find_all('span',{'class':"d-inline-block ml-0 mr-3"})
  star_fork_tag=soup.find_all("a",{'class':"Link--muted d-inline-block mr-3"})
  stars_today_tag=soup.find_all('span',{'class':'d-inline-block float-sm-right'})
  #print(len(star_fork_tag))
  for i,j in zip(programming_lang_tag,stars_today_tag):
    programming_lang.append(i.find_all("span", recursive=False)[1].text)
    stars_today.append(j.text.strip().replace('stars today',''))
  
  for k in range(len(star_fork_tag)):
    if k%2==0:
      stars.append(star_fork_tag[k].text.strip())
    else:
      fork.append(star_fork_tag[k].text.strip())
  #print(len(programming_lang_tag),len(stars),len(fork),print(len(stars_today)))
  return[programming_lang,stars,fork,stars_today]



```

```python
def final_sol():
  response=get_topics_page()
  resp1=scrape_username_topic_url(response)
  resp2=scrape_prog_fork_stars_starstoday(response)
  final_dict={
      "username":resp1[0][:20],
      "topic":resp1[1][0:20],
      "topic_desc":resp1[3][0:20],
      "url":resp1[2][0:20],
      "stars":resp2[1][:20],
      "fork":resp2[2][:20],
      "stars_today":resp2[3][:20]    
  }
  print(final_dict)

  return pd.DataFrame(final_dict)
```
